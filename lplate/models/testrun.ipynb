```python
import time
from ultralytics import YOLO
import torch
import cv2
import numpy as np
import pytesseract
```

```python
model = YOLO("yolov8n.pt")
```

```python
results = model.predict(source="./../image.png", show=True, conf=0.4)
```

```python
for result in results:
    print(result.boxes.cls , result.boxes.xyxy)
```

```python
image = cv2.imread("./../image.png")
for result in results:
    # Iterate over class IDs and bounding box coordinates
    for class_id, box_tensor in zip(result.boxes.cls, result.boxes.xyxy):
        # Convert the class ID to a regular Python integer
        class_id = int(class_id)

        # Check if the detected object is a car (assuming class ID 2 represents cars)
        if class_id == 2:
            # Convert the bounding box tensor to a list of coordinates
            box_coordinates = box_tensor.tolist()

            # Extract the bounding box coordinates
            x_min, y_min, x_max, y_max = map(int, box_coordinates)

            # Extract the region of the car
            car_region = image[y_min:y_max, x_min:x_max]

            # Apply morphological operations to enhance the license plate
            kernel = np.ones((3, 3), np.uint8)
            car_region_processed = cv2.morphologyEx(
                car_region, cv2.MORPH_CLOSE, kernel)

            # Apply OCR to the processed car region to recognize license plate text
            license_plate_text = pytesseract.image_to_string(
                car_region_processed)

            # Print the recognized license plate text
            print("License Plate:", license_plate_text)

            # Display the processed region
            cv2.imshow("Processed Car Region", car_region_processed)
```
